## changing data types
Now we will start changing apporpiate data types for each columns

1st: converting data_added to datetime

```python
df['date_added'] = pd.to_datetime(df['date_added'] , errors = 'coerce')
```

2nd: Convert duration to numeric if in case we have to extract it numbers

```python
df['duration'] =pd.to_numeric(df['duration'] , errors='coerce').astype(float)

```

```python
df['rating'] = df['rating'].astype('category')
```

```python
df.info()
```

if we take a look at the data.info() above we will se all the data types we altered are changed

now we will check for all the columns which have null values and correct it 

```python
df.director.unique()
```

```python
df.director.value_counts()
```

```python
#as director cant be fill with mode or mean so we will fill it wwith unknown
df['director'] = df['director'].fillna('Unknown')
```

```python
## 2nd we will do it with cast 
df['cast'] = df['cast'].fillna('Unknown_cast')
```

```python
#countries have a lot of missing values and we cant fill it unknown as it will manipulate data
#so we will remove country column
df.drop('country' , axis =1 , inplace = True )
```

```python
## we will replace small null values with mode
df['date_added'] = df['date_added'].fillna(df['date_added'].mode())
df['rating'] = df['rating'].fillna(df['rating'].mode())
df['duration'] = df['duration'].fillna(df['duration'].mode)
```

```python
#heatmaps for correlation matrix
#Calculate correlation matrix
corr = df.corr(numeric_only=True)

#plot the heatmap
plt.Figure(figsize=(6,6))
sns.heatmap(corr , annot=True , cmap='coolwarm' , fmt='.2f' , linewidths=0.5)
plt.title('Correlation heatmap for numerical Features')
plt.show()
```
